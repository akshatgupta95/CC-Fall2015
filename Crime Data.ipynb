```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/akshungupta/Downloads/crimes.csv', sep=',')
```

```python
df
```

```python
df2 = df[['Year']]
df2['Counts'] = 1
df2.groupby(by='Year').sum().plot(kind='bar',figsize=(20,10), title="Number of Reported Crimes by Year")
```

```python
df3 = df[['Primary Type']]
df3['Counts'] = 1
df3.groupby(by='Primary Type').sum().sort(['Counts'],ascending=False).plot(kind = "bar", figsize=(20,10), title="Number of Reported Crimes by Crime Type")
```

```python
df4 = df[['Arrest']]
df4['Counts'] = 1
df4.groupby(by='Arrest').sum().sort(['Counts'],ascending=False).plot(kind = "bar", figsize=(15,10),title = "Number of reported Crimes by Arrest")
```

```python
df5 = df[['Community Area']]
df5['Counts'] = 1
df5.groupby(by='Community Area').sum().sort(['Counts'],ascending=False).plot(kind = "bar", figsize=(30,20), title="Number of Reported Crimes by District")
```

```python
df6 = df[['Location Description']]
df6['Counts'] = 1
df6.groupby(by='Location Description').sum().sort(['Counts'],ascending=False).plot(kind = "bar", figsize=(30,20), title="Number of Reported Crimes by Location Description")
```
